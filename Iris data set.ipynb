```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris_data = load_iris()
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = pd.Categorical.from_codes(iris_data.target, iris_data.target_names)

print("üîç First 5 rows of dataset:")
print(df.head())
print("\nüìè Dataset Shape:", df.shape)
print("\nüìä Dataset Info:")
print(df.info())

print("\n‚ùì Missing Values:\n", df.isnull().sum())

print("\nüìà Statistical Summary:\n", df.describe())

print("\nüìå Species Count:")
print(df['species'].value_counts())

df.hist(figsize=(10, 6), bins=15, color='skyblue', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()

sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.suptitle("Pairplot of Iris Dataset", y=1.02, fontsize=16)
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y=df.columns[0], data=df)
plt.title(f"{df.columns[0]} by Species")
plt.show()

plt.figure(figsize=(8, 5))
sns.heatmap(df.drop('species', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()


print("""
üîπ Iris dataset has 150 rows & 5 columns (4 numeric features + species).
üîπ No missing values found.
üîπ Each species has 50 samples.
üîπ Setosa has generally smaller petal & sepal sizes, Virginica has larger.
üîπ Strong correlation between petal length & petal width.
""")
```
